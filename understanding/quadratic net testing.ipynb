### Scenario 0

In this scenario, the functions of any_of and all_of are shown to work correctly in the independendnt case.  Here hypertension, disorders of the lipid metabolism, and diabetes are all (incorrectly) independent , just as an example to show what happens when later dependencies are added. Here and in every scenario for comparison, the probabilities of different combinations of variables are computed with pomegranate probability function, the priors of all variables with the predict_proba function, and the probabilities of all the variables given evidence of hypertension and not hypertension with the predict_proba function.    Because they are independent, and they are all at half a chance, the probability of any particular combination is .5 x .5 x .5 = .125.  all_of is the chance that all the variable values are true, which is the chance of the particular combination of (hypertension, disorders of lipid metabolism, and diabetes) = .5 x .5 x .5 = .125.  any_of is 1 - the chance that none of the variables are true = 1- (no hypertension, no disorders of the lipid metabolism, no diabetes) = 1-.125 = .875.  We see that they are all computed correctly.
